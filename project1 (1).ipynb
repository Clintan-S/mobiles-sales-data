```python


mask = df[(df['Spec Score'] > 95) & (df['rating'] > 4.5)].count()

plt.scatter(df['Spec Score'], df['rating'], color='grey', alpha=0.5)
plt.scatter(df.loc[mask, 'Spec Score'], df.loc[mask, 'rating'], color='red', marker='o')
plt.title("mobile sales data")
plt.xlabel("spec score")
plt.ylabel("rating")
plt.show()
```

```python
X = df[['Spec Score']]
y = df['price']
model = LinearRegression()
model.fit(X, y)
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
predicted = model.predict(X)
print("Predicted prices:", predicted)
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, predicted, color='red', label='Regression Line')
plt.xlabel("Spec Score")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
X = df[['rating']]
y = df['price']
model = LinearRegression()
model.fit(X, y)
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
predicted = model.predict(X)
print("Predicted prices:", predicted)
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, predicted, color='red', label='Regression Line')
plt.xlabel("rating")
plt.ylabel("price")
plt.legend()
plt.show()
```
